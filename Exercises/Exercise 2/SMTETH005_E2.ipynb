```python
# import packages 
import matplotlib.pyplot as plt
import numpy as np
import math

# define constants 
k_prime = 0.1 # (d^-1) instantaneous growth rate 
P_0 = 0.1 # mmol N m^-3 concentration of nitrogen in the producers at t = 0 
N_0 = 5.0 # mmol N M^-3 concentration of nitrogen in the environment at t = 0

# generate time values to run through the function 
t = np.linspace(0,10,240) # generate an array from 0 to 10 with 240 points (one for each our of ten days)

# define the function for consumer growth
def f(t):
    return (P_0*(P_0+N_0)/(P_0+N_0*math.e**(-k_prime*(P_0+N_0)*t)))

population_values = f(t) # compute the population size at all times 

# define the function for nutrient depletion 

def g(t):
    return (P_0+N_0-population_values) # the nutrients available at any time is equal to the total nutrients (initial condition) minus that already assimilated into the population 

nutrient_values = g(t) # compute the nutrient pool at all times 

# plot the functions for the time domain
plt.plot(t, population_values, label = 'Consumer', color = 'red') # consumer growth 
plt.plot(t, nutrient_values, label = 'Nutrient', color = 'blue' ) # nutrient drawdown
plt.xlabel('Elapsed Time (days)')
plt.ylabel(r'$mmol\ N\ m^{-3}$')
plt.title('Consumer growth and nutrient depletion over time.')
plt.legend()
%config InlineBackend.figure_format = 'svg'
```

### Question 1 - Why are the curves identical in shape? 

The total amount of nitrogen in the system is conserved.

$$
T_0=P_0+N_0
$$

The total nitrogen is partitioned between the nutrient pool and the population.

$$
P+N=P_0+N_0
$$

where: 

$P$ = Population nitrogen 

$P_0$ = Initial population size (at $t = 0$)

$N$ = Available nitrogen 

$N_0$ = Initial available nitrogen

As the consumer population grows (increase in the amount of nitrogen in the consumers), the nutrient pool must decrease by the same amount. Therefore, the curves have exactly the same shape and are complementary: as the one increases, the other decreases.
